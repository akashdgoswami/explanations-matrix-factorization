```python
_results = {}
def load(name, run):
    if (run, name) not in _results:
        _results[run, name] = load_results('./results/drugbank-94x425-5to1/run{}/results_{}.pkl'.format(run, name))
    return _results[run, name]
```

```python
mmmf = 'mmmf_200steps'
stan = 'stan_nolookahead_s200w100_200steps'
```

```python
all_data = []
for name in (mmmf, stan):
    for run in range(5):
        try:
            d = load(name, run)
            all_data.append(d)
            for k, v in d.items():
                if not k.startswith('_') and v[0][0] != 500:
                    diff = 500 - v[0][0]
                    d[k] = [[n + diff,] + rest for n, *rest in v]
        except IOError:
            pass
```

```python
def merge(*results):
    d = {}
    for oth in results:
        for k, v in oth.items():
            if k not in d:
                d[k] = v
            elif not np.all(d[k] == v):
                print("disagree on {}".format(k))
    return d
```
