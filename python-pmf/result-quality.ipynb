```python
random_keys = ('random', 'bayes_random', 'mmmf_random', 'rc_random')def get_random(r):    for x in random_keys:        if x in r:            return r[x]
```

```python
diffs = defaultdict(list)for filename, results in zip(filenames, load_results(filenames)):    ns, random = rmse(get_random(results))    for k, v in results.items():        if k not in random_keys and not k.startswith('_'):            if 'bayes' in filename and not k.startswith('bayes_'):                k = 'bayes_' + k            diffs[k].append(random - rmse(v)[1])
```

```python
def auc(x, ns=None):    x = np.asarray(x)    if ns is None:        ns = np.arange(x.size)    ns = np.asarray(ns)    return ((x[:-1] + x[1:]) * np.diff(ns)).sum() / 2
```

```python
diff_aucs = {k: [auc(x) for x in v] for k, v in diffs.items()}diff_means = {k: [np.mean(x) for x in v] for k, v in diffs.items()}
```
